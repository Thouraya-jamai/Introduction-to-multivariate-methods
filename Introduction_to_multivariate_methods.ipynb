```python
import pandas as pd
```

```python
#create the data matrix and named stat
statdata = {
    'governorate': ['Tunis', 'Tunis', 'Kairouan', 'Kairouan', 'Sfax', 'Sfax', 'Gafsa', 'Gafsa'],
    'year': [2009, 2010, 2009, 2010, 2009, 2010, 2009, 2010],
    'population': [996.4, 1000.3, 558.2, 559.7, 918.5, 931.0, 335.1, 338.1]
}
stat=pd.DataFrame(statdata)
stat
```

```python
#Change the order of the columns (year, governorate, population) and display the new result
stat=stat[['year', 'governorate', 'population']]
stat
```

```python
#Create a new data matrix from stat, called stat2, by changing the row index (alphabetical number instead of numeric)
#and adding another initially empty Job variable.
#Then fill in the Employment column with the value 16.5
stat2=stat.copy()
stat2.index = [chr(i) for i in range(65, 65 + len(stat2))]
stat2['Job']=None
stat2['Employment']=16.5
stat2
```

```python
#Add another variable to stat2, named North, filled with True for the northern governorates and False otherwise.
#Then delete this variable.

northern_governorates = ['Tunis']
stat2['North'] = stat2['governorate'].isin(northern_governorates)
stat2
```

```python
#delete the variable 'North'
stat2 = stat2.drop(columns=['North'])
stat2
```

```python
import pandas as pd
```

```python
dt={
   'name' :['Ali','Mohamed','Salma','Sami','Imen'],
    'Math':[9.5,15,12,16,14],
    'Info':[14,13,11,17,7],
    'Physique':[7.5,14,13,12,8.5]
}
data=pd.DataFrame(dt)
data
```

```python
#1) Determine the line vector of the arithmetic averages of the grades of three subjects
averages=data[['Math','Info','Physique']].mean()
averages
```

```python
#2) Determine the line vector of the arithmetic ranges of the grades of three subjects
ranges=data[['Math','Info','Physique']].max() - data[["Math", "Info", "Physique"]].min()
ranges
```

```python
#3) Determine the line vector of the arithmetic medians of the grades of three subjects
medians=data[["Math", "Info", "Physique"]].median()
medians
```

```python
#4) Calculate the Y matrix of centered data and the Z matrix of centered and reduced data.
grades = data[["Math", "Info", "Physique"]]

# (Y matrix)
Y = grades - averages

# (Z matrix)
std_devs = grades.std()
Z = Y / std_devs


print("Y matrix (Centered Data):")
print(Y)

print("\nZ matrix (Centered and Reduced Data):")
print(Z)
```

```python
#5) Calculate the RX matrix of the correlations of our data matrix and the VZ matrix of the variances and covariances of Z.
import numpy as np

RX = Z.corr()

VZ = np.cov(Z.T)

print("matrix of the correlations (RX matrix):")
print(RX)

print("\nmatrix of the variances and covariances (VZ matrix):")
print(VZ)
```
